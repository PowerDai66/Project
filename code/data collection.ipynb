combine counts

```python
dems_count = pd.DataFrame(data=dems_count)
reps_count = pd.DataFrame(data=reps_count)

dems_count = dems_count[dems_count.iloc[:,1].astype(float) > 2] #choose followers who followess more than 2 politicians from a certain party
reps_count = reps_count[reps_count.iloc[:,1].astype(float) > 2]

```

```python
combined_count = dems_count
combined_count.insert(combined_count.shape[1], 'reps', 0)
combined_count.columns=['ID','dems','reps']
```

```python
for i in range(len(reps_count)):
    if any(combined_count.ID == reps_count.iloc[:,0][i]):
        current_ID = reps_count.iloc[:,0][i]
        index = combined_count[(combined_count.ID==current_ID)].index.tolist()[0]   #index in the original data frame
        combined_count.loc[index,'reps']= reps_count.iloc[:,1][i]
    else:
        new_row = pd.DataFrame([(reps_count.iloc[:,0][i],0,reps_count.iloc[:,1][i])], columns = ['ID' , 'dems', 'reps'])
        combined_count = combined_count.append(new_row, ignore_index=True)


        
```

```python
#suppress scientific notation
# reference :https://www.git2get.com/av/107633758.html
def deal_str(data):
    data = str(data)+'\t'
    return data
combined_count['ID'] = combined_count['ID'].map(deal_str)

combined_count.to_csv(r'C:\Users\ddl11\Desktop\project\accounts\combined_count.csv')
```

```python
#dems #IDs
path=r'C:\Users\ddl11\Desktop\project\accounts\account_filtering\7-8more_dems.csv'
file = pd.read_csv(path)
IDs = file['ID']
from datetime import datetime 

filtered_ID = list()
dropped_ID = list()
for ID in IDs:
    try:
        timeline = api.user_timeline(user_id=ID,count=5,tweet_mode="extended")      #5 tweets for each account
        recent_tweets = 0
        for i in range(len(timeline)):
            if datetime.now() - timeline[i].created_at <= pd.Timedelta('7 days'):#less than 2 tweets in last week, non-active account,drop
                recent_tweets+=1 
        if  recent_tweets >=2:
            filtered_ID.append(ID)
        else: 
            dropped_ID.append(ID)
    except tweepy.TweepError as e: #drop protected account
        dropped_ID.append(ID) 
    pass
```
